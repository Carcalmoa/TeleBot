### Manager Conversacional

```python
class ConversationManager():
  """
  Componente que gestiona la automatización de la conversación y la lógica
  de interacción con el backend
  """

  backend = None
  nlg = None
  nlu_component = None
  close_conversation = None
  quiere_reunion = None

  def __init__(self, nlu_component, static_patterns = False):

    self.nlu_component = nlu_component
    self.backend = Backend()
    self.nlg = NLG()

    self.static_patterns = static_patterns
    self.close_conversation = False
    self.quiere_reunion = False
    self.quiere_subir_en_bus = False

  def is_last_message(self):
    """Comprueba si el chatbot da por finalizada la conversacion"""
    return self.close_conversation

  def init_conversation(self):
    """El mensaje con el que el chatbot inicia la conversación con el usuario"""
    self.close_conversation = False
    username = None
    response = self.nlg.say_hi(username)
    if self.static_patterns:
      response += "\n" + self.nlg.say_commands()
    return response

  def _get_entity_values(self, entities, keyword, only_one):
    """Extrae la entidad en caso de que el usuario la haya introducido"""
    entity_values = [entity["value"] for entity in entities if entity["entity"] == keyword]
    if only_one:
      if len(entity_values)>0:
        return entity_values[0]
      else:
        return None
    return entity_values

  def logic(self, user_message):
    """Logica que controla el chatbot"""
    if self.quiere_reunion:
      return self._reunion_logic(user_message)
    elif self.quiere_subir_en_bus:
      return self._subir_en_bus_logic(user_message)
    else:
      return self._general_logic(user_message)

  def _general_logic(self, user_message):
    """Logica asociada a la gestion de los comandos generales"""
    intent, entities = self.nlu_component(user_message)

    # Saludos
    if intent == "greet":
      # Extrae el nombre del usuario de las entidades (de existir)
      if entities != None:
        username = self._get_entity_values(entities=entities, keyword="username", only_one=True)
      else:
        username = None
      response = self.nlg.say_hi(username)
      return response

    # Horarios
    if intent == "horario_cafeteria":
      horario_cafeteria = self.backend.get_horario_cafeteria()
      response = self.nlg.say_horario(horario_cafeteria)
      return response

    if intent == "horario_conserjeria":
      horario_conserjeria = self.backend.get_horario_conserjeria()
      response = self.nlg.say_horario(horario_conserjeria)
      return response

    if intent == "horario_comedor":
      horario_comedor = self.backend.get_horario_comedor()
      response = self.nlg.say_horario(horario_comedor)
      return response

    # Menús
    if intent == "menu_cafeteria":
      menu_cafeteria = self.backend.get_menu_cafeteria()
      response = self.nlg.say_menu_cafeteria(menu_cafeteria)
      return response

    if intent == "menu_comedor":
      menu_comedor = self.backend.get_menu_comedor()
      response = self.nlg.say_menu_comedor(menu_comedor)
      return response

    # Asociaciones
    if intent == "asociaciones":
      asociaciones = self.backend.get_asociaciones()
      response = self.nlg.say_asociaciones(asociaciones)
      return response

    # Máquina Expendedora
    if intent == "maquina_expendedora":
      articulos = self.backend.get_articulos_maquina_expendedora()
      response = self.nlg.say_articulos_maquina_expendedora(articulos)
      return response

    # Profesores
    if intent == "profesores":
      self.quiere_reunion = True
      profesores_data = self.backend.get_profesores()
      response = self.nlg.say_profesores(profesores_data)
      response += f"\n{self.nlg.say_preguntar_datos_reunion()}"
      return response

    if intent == "asignaturas":
      # Extrae la asignatura de las entidades (de existir)
      subject = self._get_entity_values(entities=entities, keyword="subject", only_one=True)
      subject = subject.upper()
      profesores = self.backend.get_profesor_x_asignatura(subject)
      response = self.nlg.say_profesor_asignatura(subject, profesores)
      return response

    if intent == "profesor_asignaturas":
      # Extrae el profesor de las entidades (de existir)
      profesor = self._get_entity_values(entities=entities, keyword="profesor", only_one=True)
      profesor = profesor.title()
      asignaturas = self.backend.get_asignatura_x_profesor(profesor)
      response = self.nlg.say_asignatura_profesor(profesor, asignaturas)
      return response

    # Buses
    if intent == "subir_en_bus":
      self.quiere_subir_en_bus = True
      response = "Dime tu ubicación"
      return response

    if intent == "bajar_en_bus":
      response = "\nEstos son los buses que pasarán proximamente por la parada de teleco: \n"
      #ID de la parada de teleco
      id_parada = "8740"
      response += self.backend.get_bus_info(id_parada)
      return response

    if intent == "bye":
      # Cambia el estado del gestor de conversacion (se cierra la comunicacion)
      self.close_conversation = True
      response = self.nlg.say_bye()
      return response

    return self.nlg.say_nounderstand()

  def _reunion_logic(self, user_message):
    """Logica asociada a la creación de una reunión"""
    user_message = user_message.replace(" y ", ",")
    datos_reunion = [pid.strip() for pid in user_message.split(",")]
    reunion_item = self.backend.check_data_reunion(datos_reunion)

    if reunion_item == None:
      self.quiere_reunion = False
      return reunion_item
    elif reunion_item.startswith("No has introducido") or reunion_item.startswith("El horario introducido"):
      self.quiere_reunion = False
      return reunion_item
    elif (reunion_item != None):
      # Cambia el estado del gestor de conversacion (se acabó la solicitud de reunión)
      self.quiere_reunion = False
      reunion_creada_id = self.backend.nueva_reunion(reunion_item)
      response = self.nlg.say_reunion_establecida(reunion_item, reunion_creada_id, self.backend.reuniones_establecidas)
      return response

  def _subir_en_bus_logic(self, user_message):
    # Cambia el estado del gestor de conversacion (se acabó la solicitud de información para subir en bus al CUVI)
    self.quiere_subir_en_bus = False
    paradas_data = self.backend.get_paradas_data("./sample_data/Paradas_Cuvi.json")
    response = f"{self.backend.get_closest_bus_stop(paradas_data, user_message)}"
    return response
```

```python
def channel(conversation_manager):

    # Chatbot inicia conversacion
    initial_sms = conversation_manager.init_conversation()
    if len(initial_sms)>0:
        print(f"Teleco Chatbot: {initial_sms} \n")

    while True:
        text = str(input('Yo: '))

        response = conversation_manager.logic(text)
        print(f"Teleco Chatbot: {response} \n")

        if conversation_manager.is_last_message():
          break
```

```python
import rasa
import asyncio

class RasaBasedNLU:

  def __init__(self, model_path = "./rasa_model", intent_threshold:float = 0.3):
    self.agent = asyncio.run(rasa.core.agent.load_agent(model_path=model_path))
    self.intent_threshold = intent_threshold


  def __call__(self, user_message):

    # Ejecutar el modelo NLU
    result = asyncio.run(self.agent.parse_message(user_message))

    # Intent
    if result["intent"]["confidence"]> self.intent_threshold:
        return result["intent"]["name"], result["entities"]

    return "not_found", None
```
